### 7.Using the second half of the Python code for Figure 9.2 - Simple Gaussian Naive Bayes Classification from Statistics, Data Mining, and Machine Learning in Astronomy, 2nd Edition as a guide,plot the decision boundaries for each classifier and dataset. What differences do you observe?



```python

# Decision boundaries for each classifier for dataset1
X1 = dataset1.X_1
X2 = dataset1.X_2
Y = dataset1.classify
X = dataset1[['X_1','X_2']]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.8, random_state=0)

sc = StandardScaler()
sc.fit(X_train)
X_train_nor = sc.transform(X_train)
X_test_nor = sc.transform(X_test)

print("Dataset 1\n\nDecision Boundary using Logistic Regression classifier")
# Fit the Logistic Regression classifier
model = LogisticRegression()
model.fit(X, Y)

# predict the classification probabilities on a grid
xlim = (-1, 40)
ylim = (-1, 40)
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 71),
                     np.linspace(ylim[0], ylim[1], 81))
Z = model.predict_proba(np.c_[xx.ravel(), yy.ravel()])
Z = Z[:, 1].reshape(xx.shape)

#y = np.zeros(200)
#y[100:] = 1

#------------------------------------------------------------
# Plot the results
fig = plotter.figure(figsize=(6, 5))
ax = fig.add_subplot(111)
ax.scatter(X['X_1'], X['X_2'], c=Y, cmap='coolwarm', zorder=2)
ax.contour(xx, yy, Z, [0.5], colors=('r'))
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
plotter.show()
```

```python
print("Dataset 1\n\nDecision Boundary using Gaussean Naive Bayes classifier")
# Fit the Gaussean Naive Bayes classifier
model = GaussianNB()
model.fit(X, Y)

# predict the classification probabilities on a grid
xlim = (-1, 40)
ylim = (-1, 40)
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 71),
                     np.linspace(ylim[0], ylim[1], 81))
Z = model.predict_proba(np.c_[xx.ravel(), yy.ravel()])
Z = Z[:, 1].reshape(xx.shape)

#y = np.zeros(200)
#y[100:] = 1

#------------------------------------------------------------
# Plot the results
fig = plotter.figure(figsize=(6, 5))
ax = fig.add_subplot(111)
ax.scatter(X['X_1'], X['X_2'], c=Y, cmap='coolwarm', zorder=2)
ax.contour(xx, yy, Z, [0.5], colors=('r'))
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
plotter.show()
```
